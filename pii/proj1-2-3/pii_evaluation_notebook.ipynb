```python
import pandas as pd
import json
import logging
import sys
import gc

# Add PII lighter protection module path
sys.path.append(r"C:\Users\ousse\Desktop\pii\pii_protection\pii_light.py")
from pii_protection.pii import PIIProtectionLayer
```

```python
# Load data
file_path = r"C:\\Users\\ousse\\Desktop\\pii\\augmented_data_partial.csv"
```

```python
# Read file in chunks to prevent memory overload
chunk_size = 20  # Adjust based on available memory
def detect_pii(text, language):
    """Function to detect PII in text using PIIProtectionLayer."""
    pii_layer = PIIProtectionLayer()
    result = pii_layer.process_document(text, language)
    return result
```

```python
# Apply PII detection in chunks
def process_dataframe_in_chunks(file_path, chunk_size):
    results = []
    for chunk in pd.read_csv(file_path, chunksize=chunk_size):
        chunk_results = []
        for index, row in chunk.iterrows():
            text = row.get("text", "")  # Adjust based on your CSV column name
            language = row.get("language", "en")  # Assuming a language column exists
            pii_result = detect_pii(text, language)
            chunk_results.append({
                "original_text": text,
                "masked_text": pii_result["masked_text"],
                "detected_entities": json.dumps(pii_result["detected_entities"]),
                "language": language
            })
          # Append processed chunk to results
        results.extend(chunk_results)
        del chunk, chunk_results  # Free memory
        gc.collect()  # Garbage collection
    
    return pd.DataFrame(results)
```

```python
# Process and display results
processed_df = process_dataframe_in_chunks(file_path, chunk_size)
import ace_tools as tools
tools.display_dataframe_to_user(name="PII Detection Results (Lighter)", dataframe=processed_df)
```
